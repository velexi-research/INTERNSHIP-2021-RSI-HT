```python
import torch
from matplotlib import pyplot as plt
import os
```

```python
results_location = '../models'
all_accuracies = []

for records_file in os.listdir(results_location):
    results = torch.load(os.path.join(results_location, records_file))
    losses = results['loss']
    plt.plot(losses)
    plt.show()
    print(results['args_dict'])
    all_accuracies.append(results['accuracies'])
```

```python
for test_accuracies in all_accuracies:
    plt.plot(test_accuracies)
    plt.show()
    print(results['args_dict'])
```
